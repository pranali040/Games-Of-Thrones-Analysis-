**IMDB RATINGS**

```python
plt.figure(figsize=(30,30))
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    plt.subplot(4,2,i+1)
    sns.barplot( x=data['No. of Episode (Season)'] , y=data['IMDb Rating'])
    plt.ylabel('IMDb Rating' , fontsize=19)
    plt.xlabel('No. of Episode' , fontsize=15)
    plt.title("Season" + str(i+1),fontsize=19)
plt.show()
```

```python
x,y = [],[]
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    x.append(i+1)
    y.append(data['IMDb Rating'].mean())
plt.figure(figsize=(15,5))    
sns.barplot(x,y)
plt.ylabel('IMDb Rating Avg.' , fontsize=19)
plt.xlabel('Season' , fontsize=15)
plt.title("Season Avg. IMDB Ratings" ,fontsize=19)
plt.show()
```

**METACRITIC RATINGS**

```python
plt.figure(figsize=(30,30))
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    plt.subplot(4,2,i+1)
    sns.barplot( x=data['No. of Episode (Season)'] , y=data['Metacritic Ratings'])
    plt.ylabel('Metacritic Ratings' , fontsize=19)
    plt.xlabel('No. of Episode' , fontsize=15)
    plt.title("Season" + str(i+1),fontsize=19)
plt.show()
```

```python
x,y = [],[]
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    x.append(i+1)
    y.append(data['Metacritic Ratings'].mean())
plt.figure(figsize=(15,5))    
sns.barplot(x,y)
plt.ylabel('Metacritic Avg. Ratings' , fontsize=19)
plt.xlabel('Season' , fontsize=15)
plt.title("Season Avg. Metacritic Ratings" ,fontsize=19)
plt.show()
```

**RUNNING TIME OF EACH EPISODE (SEASON WISE)

```python
plt.figure(figsize=(30,30))
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    plt.subplot(4,2,i+1)
    sns.barplot( x=data['No. of Episode (Season)'] , y=data['Running Time (Minutes)'])
    plt.ylabel('Running Time (Minutes)' , fontsize=19)
    plt.xlabel('No. of Episode' , fontsize=15)
    plt.title("Season" + str(i+1),fontsize=19)
plt.show()
```

**PER SEASON AVG.RUNNING TIME**

```python
x,y = [],[]
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    x.append(i+1)
    y.append(data['Running Time (Minutes)'].sum())
plt.figure(figsize=(15,5))    
sns.barplot(x,y)
plt.ylabel('Avg. Running Time (Minutes)' , fontsize=19)
plt.xlabel('Season' , fontsize=15)
plt.title("Season Running Time (Minutes)" ,fontsize=19)
plt.show()
```
